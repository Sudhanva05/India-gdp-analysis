```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
india_gdp = pd.read_csv("data/processed/india_gdp_cleaned.csv")

india_gdp.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(india_gdp["Year"], india_gdp["GDP"], marker="o")

plt.title("India GDP Growth")
plt.xlabel("Year")
plt.ylabel("GDP (US$)")
plt.grid(True)

plt.show()
```

```python
india_gdp["GDP_Growth_%"] = india_gdp["GDP"].pct_change()*100

india_gdp.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(india_gdp["Year"], india_gdp["GDP_Growth_%"], marker="o")

plt.title("India GDP Growth Rate (%)")
plt.xlabel("Year")
plt.ylabel("GDP Growth %")

plt.grid(True)

plt.show()
```

```python
import seaborn as sns

# compute top 10 years by GDP growth percentage
top_growth = india_gdp.sort_values("GDP_Growth_%", ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(x="Year", y="GDP_Growth_%", data=top_growth)

plt.title("Top 10 GDP in India")
plt.ylabel("Growth %")
plt.xlabel("Year")

plt.show()
```
