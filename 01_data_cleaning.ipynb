```python
import pandas as pd

df = pd.read_csv(
    "/home/student/india-gdp-analysis/data/raw/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_32.csv",
    skiprows=4
)

df.head()
```

```python
india = df[df["Country Name"] == "India"]

india.head()
```

```python
india_gdp = india.melt(
    id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code"],
    var_name = "Year",
    value_name = "GDP"
)

india_gdp.head()
```

```python
india_gdp = india_gdp[["Year","GDP"]]
india_gdp = india_gdp.dropna()

india_gdp["Year"] = india_gdp["Year"].astype(int)

india_gdp.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(12,6))

plt.plot(india_gdp["GDP"], india_gdp["Year"], marker="o")

plt.title("GDP Analysis of India")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.grid(True)
plt.show()
```

```python
india = df[df["Country Name"] == "India"]

india_gdp = india.melt(
    id_vars=["Country Name","Country Code","Indicator Name","Indicator Code"],
    var_name="Year",
    value_name="GDP"
)

# keep only real years
india_gdp = india_gdp[india_gdp["Year"].str.isnumeric()]

india_gdp["Year"] = india_gdp["Year"].astype(int)

india_gdp = india_gdp.dropna()

india_gdp = india_gdp[["Year","GDP"]]

india_gdp.head()
```

```python
india_gdp["GDP_Growth_%"] = india_gdp["GDP"].pct_change()*100
```

```python
plt.figure(figsize=(12,6))

plt.plot(india_gdp["Year"], india_gdp["GDP_Growth_%"], marker="o")

plt.title("India GDP Growth Rate (%)")
plt.xlabel("Year")
plt.ylabel("GDP Growth %")
plt.grid(True)

plt.show()
```

```python
top_growth = india_gdp.sort_values("GDP_Growth_%", ascending=False).head(10)

top_growth
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))

sns.barplot(x="Year", y="GDP_Growth_%", data=top_growth)

plt.title("Top 10 GDP in India")
plt.ylabel("Growth %")
plt.xlabel("Year")

plt.show()
```
