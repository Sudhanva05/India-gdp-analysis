```python
import pandas as pd

df = pd.read_csv(
    "/home/student/india-gdp-analysis/data/raw/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_32.csv",
    skiprows=4
)

df.head()
```

```python
india = df[df["Country Name"] == "India"]

india.head()
```

```python
india_gdp = india.melt(
    id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code"],
    var_name = "Year",
    value_name = "GDP"
)

india_gdp.head()
```

```python
india_gdp = india_gdp[["Year","GDP"]]
india_gdp = india_gdp.dropna()

india_gdp["Year"] = india_gdp["Year"].astype(int)

india_gdp.head()
```

```python
india_gdp.to_csv("data/processed/india_gdp_cleaned.csv", index=False)
```
